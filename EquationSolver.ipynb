```python
import math
```

PARAMETRI:

```python
N = [5762587, 5762398] #ne ho scritto uno a caso
P = 0.242
```

**f(x) = P / sqrt(x) - log10(N * x)** e' una funzione definita sull'intervallo delle x strettamente positive (0, +inf).<br>Quindi si deve trovare un intervallo chiuso \[a, b\] tale che **f(a) * f(b) < 0**.<br>Definisco prima la funzione f:

```python
def f(x, N):
    return P/math.sqrt(x) - math.log10(N * x)
```

Trovo l'intervallo in cui cade la soluzione:

```python
def estremi(N):
    a = 0.000001
    b = 1.0

    while(f(b, N) > 0.0):
        a = b
        b += 1 
    
    return a, b
    
```

Ora che abbiamo l'intervallo, applichiamo l'algoritmo della bisezione per trovare la soluzione:

```python
def approx(a, b, N):
    M = (a + b) / 2.0
    while(abs(f(M, N)) > 0.0000001):
        if(f(M, N) == 0.0): 
            return M
        elif(f(M, N) > 0.0): 
            a = M
        else: 
            b = M
    
        M = (a + b)/ 2.0
    return M

x = []

for n in N:
    a, b = estremi(n)
    x.append(approx(a, b, n))
    
def stampa(P, n, x):
    print("l'equazione: \n\t\t   {:.3f}\n\t\t---------- = log({:.0f} * x)\n\t\t  sqrt(x)\n\nha come soluzione:".format(P, n))
    print("\t\tx = {:.10f}\n\n[ errore ≈ {:.10f} ]".format(x, abs(f(x, n)-0.0)))
    print('\n======================================================================\n')
    
for i in range(0, len(x)):
    stampa(P, N[i], x[i])
```
